```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cosine_decay(learning_rate,global_step,decay_steps,alpha=0.0):
    global_step = min(global_step, decay_steps)
    cosine_decay = 0.5 * (1 + np.cos(np.pi * global_step / decay_steps))
    decayed = (1 - alpha) * cosine_decay + alpha
    decayed_learning_rate = learning_rate * decayed
    return decayed_learning_rate
```

```python
ee = np.arange(300)
lr = [cosine_decay(0.01,global_step = e % 50,decay_steps = 50, alpha = 0.001) for e in ee]
lr = [ cosine_decay(0.01,global_step = e % 50,decay_steps = 50, alpha = 0.001)]
```

```python
plt.plot(ee,lr)
```
